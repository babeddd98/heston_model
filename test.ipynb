```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
from arch import arch_model
plt.rcParams.update({'font.size': 20})
```

```python
# ticker1 = "BTC-USD"
ticker1 = "BZ=F"

data = yf.download(ticker1,
                   start=datetime(2000, 1, 1),
                   end=datetime(2023, 1, 1))

data["Returns"] = 100 * (data["Close"].pct_change())
x = data["Returns"].dropna()


plt.subplot(211)
plt.plot(data["Close"], linewidth=0.7,
         color="#5b5b5b",
         label="ETH-USD")
plt.grid(axis='y', linestyle='dotted', color="blue")
plt.legend()

plt.subplot(212)
plt.plot(x, linewidth=0.7,
         color="#5b5b5b",
         label="volatilit")
plt.legend()
plt.grid(axis='y', linestyle='dotted', color="blue")
```

```python
# plt.hist(x, bins=300, color='grey')
# plt.grid()
import seaborn as sns
import scipy.stats as stats
# Plot the histogram of the standardized residuals
plt.figure(figsize=(7,4))
sns.distplot(x, norm_hist=True, fit=stats.t, bins=50, color='blue')
plt.legend(('cac40','Loi de Student'))
plt.show()
```

```python
# plt.hist(x, bins=300, color='grey')
# plt.grid()
import seaborn as sns
import scipy.stats as stats
# Plot the histogram of the standardized residuals
plt.figure(figsize=(7,4))
sns.distplot(x, norm_hist=True, fit=stats.norm, bins=50, color='blue')
plt.legend(('cac40','Loi normale'))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Parameters for the t-distribution
df = 10  # degrees of freedom

# Generate x values
x = np.linspace(-5, 5, 500)

# Compute the PDF of the t-distribution
t_pdf1 = t.pdf(x, 1)
t_pdf2 = t.pdf(x, 2)
t_pdf3 = t.pdf(x, 10)

# Compute the PDF of the normal distribution
mean = 0
std = 1
normal_pdf = norm.pdf(x, mean, std)

# Plot the PDFs
plt.plot(x, t_pdf1, label="Student, k=1", color="blue")
plt.plot(x, t_pdf2, label="Student, k=2", color="purple")
plt.plot(x, t_pdf3, label="Student, k=10", color="red")
plt.plot(x, normal_pdf, label="N(0,1)", color="#00FF1F")
# plt.title("Student's t-distribution and Normal distribution")
# plt.xlabel("x")
# plt.ylabel("Densit de probabilit")
plt.grid(True)
plt.legend()
plt.show()
```
